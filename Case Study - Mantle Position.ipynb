##Case Study - Mantle Position

The figure below shows the mantle set point over the last 6 months. The set point of new mantles starts at ~5%. As the mantle wears down, the set point is slowly increased until the full mantle is worn out (typically around a set point of ~75%). At that point the worn mantle is removed and replaced with a new mantle and the set point starts over at ~5%. The below figure shows the set point trend line for 9 different mantles.

```
#import modules
import pandas as pd
import plotly.graph_objects as go

#read in data
df = pd.read_csv('/dbfs/mnt/adlsdevstoacc03/training/input/mantle_setpoint.csv')

#change type of date column to datetime
df["datetime"] = pd.to_datetime(df["datetime"],errors='coerce')

#plot data
fig = go.Figure(data=go.Scatter(x=df.datetime, y=df['setpoint']))
fig.update_layout(title_text='Morenci Crusher 2 Mantle Set Point September 2019 - March 2020',
                  xaxis_title="Date",
                  yaxis_title="Mantle Set Point (%)",
                  xaxis_rangeslider_visible=True)
fig.show()
```
