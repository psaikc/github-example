##Case Study

The figure below shows the amperage of dual motors on the apron feeder (left side and right side) feeding the primary crusher. Since the two motors are connected to the same solid shaft, it is expected that the amperage of the two motors should be the same. Damage could result if the two motors pull different amperages for long periods of time. We are told that the motors failed in late September 2019 and then were replaced; this is evident in the figure where the two trend lines start to overlap each other.

```
#import modules
import pandas as pd
import plotly.graph_objects as go

#read in data
df = pd.read_csv('/dbfs/mnt/adlsdevstoacc03/training/input/apronfdr_amps.csv')

#plot data
fig = go.Figure()

# Add trace for motor 1
fig.add_trace(go.Scatter(
    x=df.datetime,
    y=df['motor1'],
    name="motor1", #visible on mouseover
    yaxis="y" #keep track of multiple axes
))

# Add trace for motor 2
fig.add_trace(go.Scatter(
    x=df.datetime,
    y=df['motor2'],
    name="motor2", 
    yaxis="y" 
))

fig.update_layout(title_text='Apron Feeder Motor Amps September 2019 - December 2019',
                  xaxis_title="Date",
                  yaxis_title="Amperage",
                  xaxis_rangeslider_visible=True)
fig.show()
```
